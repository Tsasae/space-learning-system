```python
import time
import pandas as pd
import plotly.express as px

os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = \
    "/home/tsatsral/Documents/lms-websystem /credentials/bigquery-key.json"

client = bigquery.Client(project="lunar-lms-bigquery")

# ===== ТУРШИЛТ =====
sizes = [1000, 10000, 50000, 100000, 572137]
results = []

for size in sizes:
    print(f"\n🔄 {size:,} бичлэг туршиж байна...")
    
    # BigQuery хурд
    start = time.time()
    query = f"""
    SELECT latitude, longitude, bright_ti4, confidence, acq_date
    FROM `bigquery-public-data.nasa_wildfire.past_week`
    LIMIT {size}
    """
    df = client.query(query).to_dataframe()
    bq_time = round(time.time() - start, 3)
    
    # Pandas боловсруулалтын хурд
    start = time.time()
    df_high = df[df['confidence'] == 'high']
    avg_bright = df_high['bright_ti4'].mean()
    pandas_time = round(time.time() - start, 3)
    
    results.append({
        'Өгөгдлийн хэмжээ': f'{size:,}',
        'BigQuery татах (сек)': bq_time,
        'Pandas боловсруулах (сек)': pandas_time,
        'Нийт (сек)': round(bq_time + pandas_time, 3)
    })
    print(f"  BigQuery: {bq_time}с | Pandas: {pandas_time}с")

# Хүснэгт
result_df = pd.DataFrame(results)
print("\n📊 ҮНЭЛГЭЭНИЙ ХҮСНЭГТ:")
print(result_df.to_string(index=False))

# График
fig = px.bar(result_df,
    x='Өгөгдлийн хэмжээ',
    y=['BigQuery татах (сек)', 'Pandas боловсруулах (сек)'],
    title='🔥 NASA Wildfire — BigQuery vs Pandas Хурдны Харьцуулалт\n(572,137 бичлэг)',
    barmode='group',
    color_discrete_map={
        'BigQuery татах (сек)': '#4285F4',
        'Pandas боловсруулах (сек)': '#EA4335'
    }
)
fig.update_layout(
    xaxis_title='Өгөгдлийн хэмжээ',
    yaxis_title='Хугацаа (секунд)',
    height=500
)
fig.show()
print("✅ Туршилт дууслаа!")
```

```python
# Dataset үүсгэх
from google.cloud import bigquery

dataset_id = "lunar-lms-bigquery.wildfire_dataset"
dataset = bigquery.Dataset(dataset_id)
dataset.location = "US"

dataset = client.create_dataset(dataset, exists_ok=True)
print(f"✅ Dataset үүслээ: {dataset_id}")

# Дараа нь ML загвар сургах
ml_query = """
CREATE OR REPLACE MODEL `lunar-lms-bigquery.wildfire_dataset.confidence_model`
OPTIONS(model_type='logistic_reg') AS
SELECT
  latitude,
  longitude,
  bright_ti4,
  IF(confidence = 'high', 1, 0) as label
FROM `bigquery-public-data.nasa_wildfire.past_week`
"""
print("🔄 Cloud дээр ML загвар сургаж байна...")
client.query(ml_query).result()
print("✅ Cloud дээр ML загвар сургагдлаа!")
```

```python
# Загварын үнэлгээ
eval_query = """
SELECT *
FROM ML.EVALUATE(MODEL `lunar-lms-bigquery.wildfire_dataset.confidence_model`,
  (SELECT
    latitude,
    longitude,
    bright_ti4,
    IF(confidence = 'high', 1, 0) as label
  FROM `bigquery-public-data.nasa_wildfire.past_week`))
"""
print("📊 Загварын үнэлгээ:")
eval_df = client.query(eval_query).to_dataframe()
print(eval_df.to_string(index=False))

# Урьдчилан таах
predict_query = """
SELECT *
FROM ML.PREDICT(MODEL `lunar-lms-bigquery.wildfire_dataset.confidence_model`,
  (SELECT latitude, longitude, bright_ti4
   FROM `bigquery-public-data.nasa_wildfire.past_week`
   LIMIT 10))
"""
print("\n🔮 Урьдчилан таах үр дүн (10 бичлэг):")
pred_df = client.query(predict_query).to_dataframe()
print(pred_df[['latitude','longitude','predicted_label']].to_string(index=False))
print("\n✅ BigQuery ML бүрэн ажиллаж байна!")
```
