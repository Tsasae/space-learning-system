```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = \
    "/home/tsatsral/Documents/lms-websystem /credentials/bigquery-key.json"

from google.cloud import bigquery
import pandas as pd
import plotly.express as px
import time

client = bigquery.Client(project="lunar-lms-bigquery")

# 9.6M бичлэгийн статистик
print("🛰️ Landsat 9.6M бичлэг татаж байна...")
start = time.time()

query = """
SELECT spacecraft_id,
  COUNT(*) as scene_count,
  ROUND(AVG(cloud_cover), 2) as avg_cloud,
  MIN(date_acquired) as first_date,
  MAX(date_acquired) as last_date
FROM `bigquery-public-data.cloud_storage_geo_index.landsat_index`
WHERE cloud_cover IS NOT NULL
GROUP BY spacecraft_id
ORDER BY scene_count DESC
"""

df = client.query(query).to_dataframe()
elapsed = time.time() - start
print(f"✅ {elapsed:.2f} секундэд дууслаа!")
print(df)

# График
fig = px.bar(df,
    x='spacecraft_id',
    y='scene_count',
    color='avg_cloud',
    title=f'🛰️ Landsat — Дагуул бүрийн зургийн тоо (Нийт: 9,608,135)',
    labels={'scene_count': 'Зургийн тоо', 'spacecraft_id': 'Дагуул'},
    color_continuous_scale='Blues'
)
fig.show()
```
